```python
import pandas as pd
import requests
```

```python
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/136.0.0.0 Safari/537.36"}
url="https://candidaturaspoderjudicial.ine.mx/cycc/documentos/json/ministrosSupremaCorte.json"
```

```python
r=requests.get(url, headers=headers)
print(r.status_code)
data = r.json()
```

```python
print("Número de candidatos: ",len(data["candidatos"]))
```

```python
nombre=[]
sexo=[]
fecha_nacimiento=[]
descripcion=[]
num_boleta=[]
vision_jurisdiccional=[]
vision_impjusticia=[]
propuesta1=[]
propuesta2=[]
propuesta3=[]
url_cv=[]
url_foto=[]
estatus_candidatura=[]

for candidato in data["candidatos"]:
    num_boleta.append(candidato.get("numListaBoleta"))
    nombre.append(candidato.get("nombreCandidato"))
    sexo.append(candidato.get("sexo"))
    fecha_nacimiento.append(candidato.get("fechaNacimiento"))
    estatus_candidatura.append(candidato.get("estatusVigente"))
    descripcion.append(candidato.get("descripcionCandidato"))
    vision_jurisdiccional.append(candidato.get("visionJurisdiccional"))
    vision_impjusticia.append(candidato.get("visionImparticionJusticia"))
    propuesta1.append(candidato.get("propuesta1"))
    propuesta2.append(candidato.get("propuesta2"))
    propuesta3.append(candidato.get("propuesta3"))
    cv = candidato.get("descripcionHLC")
    url_cv.append("https://candidaturaspoderjudicial.ine.mx/cycc/documentos/cv/" + cv if cv else None)
    foto=candidato.get("urlFoto")
    if foto:
        if "/media/cycc/img/fotocandidato/" in foto:
            foto = foto.replace("/media/cycc/img/fotocandidato/", "")
        url_foto.append("https://candidaturaspoderjudicial.ine.mx/cycc/img/fotocandidato/" + foto)
    else:
        url_foto.append(None)
```

```python
#Tabla
candidatas = pd.DataFrame({
    "num_boleta": num_boleta,
    "nombre": nombre,
    "sexo": sexo,
    "fecha_nacimiento": fecha_nacimiento,
    "estatus_candidatura": estatus_candidatura,
    "url_cv": url_cv,
    "url_foto": url_foto,
    "descripcion": descripcion,
    "vision_jurisdiccional": vision_jurisdiccional,
    "vision_impjusticia": vision_impjusticia,
    "propuesta1": propuesta1,
    "propuesta2": propuesta2,
    "propuesta3": propuesta3
})
#Ordenar por num_boleta
candidatas = candidatas.sort_values(by="num_boleta")
#si estatus_candidatura es 5, es vigente, si no, es cancelada
candidatas["estatus_candidatura"] = candidatas["estatus_candidatura"].replace({5: "vigente", 20: "cancelada"})
candidatas
```
